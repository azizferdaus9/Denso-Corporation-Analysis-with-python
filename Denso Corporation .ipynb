```python
#importing all the needed packages for data analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#location of the IBM data 
path = '/content/drive/MyDrive/Colab Notebooks/assignment/data/'
#location of all the attribute files
attribute5 = path + 'USPTO_2012_5_proj_attributes_NEW.csv'
```

```python
attribute_df5 = pd.read_csv(attribute5)
final_attribute = attribute_df5.drop(columns=['Unnamed: 10'])
print('total shape of attributes {}'.format(final_attribute.shape))
```

```python
df = final_attribute
print(df.shape)
df.tail()
```

```python
total = df.isnull().sum().sort_values(ascending=False)
percent = ((df.isnull().sum()/df.isnull().count())*100).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data
```

```python
df['Turnover_lay'].fillna((df['Turnover_lay'].mean()), inplace=True)
df['Turnover_2012'].fillna((df['Turnover_2012'].mean()), inplace=True)
df['Total_assets_2012'].fillna((df['Total_assets_2012'].mean()), inplace=True)
df['Employees_2012'].fillna((df['Employees_2012'].mean()), inplace=True)
df['R&D_2012'].fillna((df['R&D_2012'].mean()), inplace=True)
df['Country_code'].fillna((df['Country_code'].mean()), inplace=True)
```

```python
total = df.isnull().sum().sort_values(ascending=False)
percent = ((df.isnull().sum()/df.isnull().count())*100).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)
fig.suptitle('figure 4. data destribution of categorical columns')


# Patent_industry
sns.barplot(ax=axes[0], x=df.Patent_industry.value_counts().index, y=df.Patent_industry.value_counts().values)
axes[0].set_title('Patent_industry')


# University
sns.barplot(ax=axes[1], x=df.University.value_counts().index, y=df.University.value_counts().values)
axes[1].set_title('University')

# Country_code
sns.barplot(ax=axes[2], x=df.Country_code.value_counts().index, y=df.Country_code.value_counts().values)
axes[2].set_title('Country_code');
```

```python
keys = [company for company, df in df.groupby(['ID'])]

plt.bar(keys,df.groupby(['ID']).sum()['Total_assets_2012'])
plt.ylabel('Total asset in 2012')
plt.xlabel('Company Name ')
plt.xticks(keys, rotation='vertical', size=8)
plt.title('Figure 7. Total asset by each comapany in 2012')
plt.show()
```

```python

keys = [company for company, df in df.groupby(['ID'])]

plt.bar(keys,df.groupby(['ID']).sum()['R&D_2012'])
plt.ylabel('R&D in 2012')
plt.xlabel('Company Name ')
plt.xticks(keys, rotation='vertical', size=8)
plt.title('Figure 8. Total R&D by each comapany')
plt.show()
```

```python

keys = [company for company, df in df.groupby(['ID'])]

plt.bar(keys,df.groupby(['ID']).sum()['Employees_2012'])
plt.ylabel('Number of employees in 2012')
plt.xlabel('Company Name ')
plt.xticks(keys, rotation='vertical', size=8)
plt.title('Figure 9. Number of employees by each comapany')
plt.show()
```

```python
data = df[['Patent_count','Turnover_lay','Turnover_2012','Total_assets_2012', 'Employees_2012','R&D_2012']]
plt.figure(figsize=(16, 6))
corr = data.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
heatmap = sns.heatmap(data.corr(),mask=mask, vmin=-1, vmax=1, annot=True)
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12);
plt.title('Figure 10. Correlation between variables')

plt.show()
```

```python
total_asset = df.groupby('ID').mean()['Total_assets_2012']
keys = [company for company, df in df.groupby(['ID'])]

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(keys,df.groupby(['ID']).sum()['Turnover_2012'], color='g')
ax2.plot(keys, total_asset, color='b')

ax1.set_xlabel('Company Name')
ax1.set_ylabel('Turnover in 2012', color='g')
ax2.set_ylabel('total asset', color='b')
ax1.set_xticklabels(keys, rotation='vertical', size=8)
plt.title('\n Figure 11. Turnover and total asset trend for each companies\n')

fig.show()
```

```python
Employees_2012 = df.groupby('ID').mean()['Employees_2012']
keys = [company for company, df in df.groupby(['ID'])]

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(keys,df.groupby(['ID']).sum()['Turnover_2012'], color='g')
ax2.plot(keys, total_asset, color='b')

ax1.set_xlabel('Company Name')
ax1.set_ylabel('Turnover in 2012', color='g')
ax2.set_ylabel('Employee number in 2012', color='b')
ax1.set_xticklabels(keys, rotation='vertical', size=8)

fig.show()
```
